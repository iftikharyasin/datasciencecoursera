```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}

try:
    response = requests.get(url, headers=headers, timeout=10)
    response.raise_for_status()
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    tables = soup.find_all('table')
    target_table = None
    
    for table in tables:
        if 'Tesla Quarterly Revenue' in str(table):
            target_table = table
            break
    
    if target_table:
        tesla_revenue = pd.read_html(str(target_table))[0]
        tesla_revenue.columns = ['Date', 'Revenue']
        
        tesla_revenue['Revenue'] = (
            tesla_revenue['Revenue']
            .str.replace('[\$,]', '', regex=True)
            .astype(float)
        )
        
        tesla_revenue = tesla_revenue.dropna()
        print("Most recent Tesla quarterly revenue:")
        display(tesla_revenue.head())
        
    else:
        print("Could not find revenue table - website structure may have changed")

except Exception as e:
    print(f"Error: {e}")
    print("Try checking the URL or website structure")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

def make_graph(stock_data, title):
    plt.figure(figsize=(14, 7))
    plt.plot(stock_data.index, stock_data['Close'], label='Closing Price', color='blue')
    plt.title(title, fontsize=16)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Price ($)', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend(fontsize=12)
    plt.show()

make_graph(tesla_data, "Tesla (TSLA) Stock Price History (2010-Present)")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

def make_graph(stock_data, title):
    plt.figure(figsize=(14, 7))
    plt.plot(stock_data.index, stock_data['Close'], 
             label='Closing Price', 
             color='red',
             linewidth=1.5)
    
    plt.title(title, fontsize=16, pad=20)
    plt.xlabel('Year', fontsize=14)
    plt.ylabel('Price ($)', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.5)
    
    plt.axvspan('2020-01-01', '2021-06-30', 
                color='orange', alpha=0.2, 
                label='Meme Stock Surge')
    
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.show()

make_graph(gme_data, "GameStop (GME) Stock Price History (2002-Present)")
```
